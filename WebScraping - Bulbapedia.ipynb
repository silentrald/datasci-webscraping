# Bulbapedia

Get the list of the next generations of Pokemons. Make sure you get the following information:

- kdex
- ndex
- name
- types
- URL to the Pokemon's wiki page

```python
# Imports
import requests
from bs4 import BeautifulSoup
import json
```

```python
# URL
BASE_URL = 'https://bulbapedia.bulbagarden.net'
URL = BASE_URL + '/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number'
```

```python
page = requests.get(URL)
# page.content
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# Get main content <div>
poke_content = soup.find(id='mw-content-text')

# Get all <table> elements
poke_tables = poke_content.find_all('table')

# Get the first generation table
gen1_list = poke_tables[1]

# gen1_list.contents
```

```python
# Generate the json file
info_start = 3
gen1_json = []

for i in range(info_start, len(gen1_list.contents), 2):
    poke_info=gen1_list.contents[i]
    
    kdex = poke_info.contents[1].text.strip()
    ndex = poke_info.contents[3].text.strip()
    type1=poke_info.contents[9].text.strip()

    name_link = poke_info.contents[7]
    name = name_link.text.strip()
    link = BASE_URL + name_link.find('a').attrs['href']
    
    pokemon = {
        "kdex": kdex,
        "ndex": ndex,
        "name": name,
        "type1": type1
    }
    
    if len(poke_info.contents) > 10:
        type2 = poke_info.contents[11].text.strip()
        pokemon['type2'] = type2
    
    pokemon['link'] = link
    
    gen1_json.append(pokemon)

gen1_json
```

```python
# Write to file
with open('gen1_pokemon.json', 'w') as fp:
    fp.write(json.dumps(gen1_json, indent=4))
    fp.close()
```
